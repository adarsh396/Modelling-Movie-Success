```python
# Extracting the genre list for each movie
def extract_genres(genre_str):
    try:
        genres = ast.literal_eval(genre_str)
        return [genre['name'] for genre in genres]
    except:
        return []

movies['genres_list'] = movies['genres'].apply(extract_genres)

# Exploding the genres_list to create a separate row for each genre
movies_exploded = movies.explode('genres_list')

# Group by genre and calculate the average revenue for each genre
genre_revenue = movies_exploded.groupby('genres_list')['revenue'].mean().sort_values(ascending=False)

# Plot the average revenue by genre
plt.figure(figsize=(10,6))
genre_revenue.plot(kind='bar', color='skyblue')
plt.title("Average Revenue by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Revenue")
plt.xticks(rotation=90)
plt.show()

# Display the top genres by average revenue
print(genre_revenue.head())
```

# Step 4: Statistical Analysis and Hypothesis Testing

* # 4.1: Hypothesis Testing â€“ Correlation Test (Pearsonâ€™s Correlation)

```python
from scipy.stats import pearsonr

# Perform Pearson's correlation test between budget and revenue
corr_budget_revenue, p_value = pearsonr(movies['budget'], movies['revenue'])

print(f"Pearson's Correlation between Budget and Revenue: {corr_budget_revenue}")
print(f"P-value: {p_value}")

# If the p-value is less than 0.05, we can conclude that the relationship is statistically significant
if p_value < 0.05:
    print("The relationship between budget and revenue is statistically significant.")
else:
    print("The relationship between budget and revenue is not statistically significant.")
```

* # 4.2: Linear Regression Model â€“ Predicting Revenue

* 4.2.1: Data Preprocessing for Regression Model

```python
import ast

# Function to extract genres from the JSON-like string
def extract_genres(genre_str):
    try:
        genres = ast.literal_eval(genre_str)  # Convert the string to a list of dicts
        return [genre['name'] for genre in genres]  # Extract genre names
    except:
        return []

# Apply the function to create the genres list
movies['genres_list'] = movies['genres'].apply(extract_genres)

# Now we will explode the genres_list to create a separate row for each genre
movies_exploded = movies.explode('genres_list')

# Create dummy variables for genres
genres_dummies = pd.get_dummies(movies_exploded['genres_list'], prefix='genre')

# Group by movie_id and aggregate the dummy variables
genres_dummies = genres_dummies.groupby(movies_exploded['id']).max()

# Add these genre dummies to the movies dataset
movies = movies.join(genres_dummies, on='id')

# Fill missing values for runtime and vote_average
movies.fillna({'runtime': movies['runtime'].median(), 'vote_average': movies['vote_average'].median()}, inplace=True)

# Select features for the regression model
X = movies[['budget', 'runtime', 'vote_average'] + list(genres_dummies.columns)]
y = movies['revenue']

# Feature scaling (normalize the features)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train a Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print(f"R-squared (Train): {train_score}")
print(f"R-squared (Test): {test_score}")

# Display the model coefficients
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coefficients)
```

* 4.2.2: Model Interpretation

```python
# Display the model coefficients
coefficients = pd.DataFrame(model.coef_, X.columns, columns=['Coefficient'])
print(coefficients)
```

# Step 5: Model Evaluation and Improvement

* # 5.1: Evaluating the Model's Performance

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Predict the values for the test set
y_pred = model.predict(X_test)

# Calculate MAE, MSE, and RMSE
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

* # 5.2: Model Diagnostics â€“ Residuals Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate residuals
residuals = y_test - y_pred

# Plot residuals to check for any patterns
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, color='blue')
plt.title("Residuals Distribution")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()

# Plot a scatter plot of residuals vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.5, color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.title("Residuals vs Predicted Values")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```

* # 6.1: Regularization â€“ Ridge and Lasso Regression

* 6.1.1: Ridge Regression 

```python
from sklearn.linear_model import Ridge

# Initialize Ridge regression model with alpha (regularization strength)
ridge_model = Ridge(alpha=1.0)  # You can experiment with different alpha values

# Train the model
ridge_model.fit(X_train, y_train)

# Evaluate the model
ridge_train_score = ridge_model.score(X_train, y_train)
ridge_test_score = ridge_model.score(X_test, y_test)

# Print the evaluation results
print(f"Ridge R-squared (Train): {ridge_train_score}")
print(f"Ridge R-squared (Test): {ridge_test_score}")
```

* 6.1.2: Lasso Regression

```python
from sklearn.linear_model import Lasso

# Initialize Lasso regression model with alpha (regularization strength)
lasso_model = Lasso(alpha=0.01)  # You can experiment with different alpha values

# Train the model
lasso_model.fit(X_train, y_train)

# Evaluate the model
lasso_train_score = lasso_model.score(X_train, y_train)
lasso_test_score = lasso_model.score(X_test, y_test)

# Print the evaluation results
print(f"Lasso R-squared (Train): {lasso_train_score}")
print(f"Lasso R-squared (Test): {lasso_test_score}")
```

* # 6.2: Non-Linear Models â€“ Random Forest Regressor

* 6.2.1: Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Evaluate the model
rf_train_score = rf_model.score(X_train, y_train)
rf_test_score = rf_model.score(X_test, y_test)

# Print the evaluation results
print(f"Random Forest R-squared (Train): {rf_train_score}")
print(f"Random Forest R-squared (Test): {rf_test_score}")
```

* # 6.3: Hyperparameter Tuning â€“ Grid Search

```python
from sklearn.model_selection import GridSearchCV

# Set up parameter grid for Ridge regression
ridge_param_grid = {'alpha': [0.1, 1.0, 10.0, 100.0]}
ridge_grid_search = GridSearchCV(Ridge(), ridge_param_grid, cv=5)
ridge_grid_search.fit(X_train, y_train)
print(f"Best parameters for Ridge: {ridge_grid_search.best_params_}")
print(f"Best score for Ridge: {ridge_grid_search.best_score_}")

# Set up parameter grid for Lasso regression
lasso_param_grid = {'alpha': [0.01, 0.1, 1.0, 10.0]}
lasso_grid_search = GridSearchCV(Lasso(), lasso_param_grid, cv=5)
lasso_grid_search.fit(X_train, y_train)
print(f"Best parameters for Lasso: {lasso_grid_search.best_params_}")
print(f"Best score for Lasso: {lasso_grid_search.best_score_}")

# Set up parameter grid for Random Forest
rf_param_grid = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20, 30]}
rf_grid_search = GridSearchCV(RandomForestRegressor(), rf_param_grid, cv=5)
rf_grid_search.fit(X_train, y_train)
print(f"Best parameters for Random Forest: {rf_grid_search.best_params_}")
print(f"Best score for Random Forest: {rf_grid_search.best_score_}")
```
